```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/sqft_price.csv')
x_train = data['sqft'].values
y_train = data['price'].values
```

```python
plt.scatter(x_train,y_train)
plt.xlabel('Sq. Feet')
plt.ylabel('Price')
plt.show()
```

```python
def cost_function(x, y, w, b):
  m = len(x)
  total_cost = 0

  for i in range(m):
    f = w * x[i] + b
    cost = (f - y[i]) ** 2
    total_cost += cost

  cost_fun = (1/(2*m)) * total_cost
  return cost_fun
```

```python
def gradient_desc(w, b, x, y):
    m = len(x)
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f = w * x[i] + b
        dj_dw += (f - y[i]) * x[i]
        dj_db += (f - y[i])

    dj_dw = (1/m) * dj_dw
    dj_db = (1/m) * dj_db

    return dj_dw, dj_db
```

```python
def gradient_desc_func(x, y, alpha, iter):
  w = 0
  b = 0

  for i in range(iter):
    dj_dw, dj_db = gradient_desc(w,b,x,y)
    w = w - alpha*dj_dw
    b = b - alpha*dj_db
    print(f"Iter: {i}, Cost: {cost_function(x_train,y_train,w,b)}")

  return w,b
```

```python
alpha = 0.000000001
iter = 10000

final_w, final_b = gradient_desc_func(x_train, y_train ,alpha,iter)
print(f"final_w: {final_w:.4f}, final_b: {final_b:.4f}")
```

```python
plt.scatter(x_train, y_train,label='Actual Data')
plt.plot(x_train, final_w * x_train + final_b, color='red',label='Regression line')
plt.xlabel('Sq Feet',)
plt.ylabel('Price')
plt.title('House Price prediction')
plt.legend()
plt.show()
```
